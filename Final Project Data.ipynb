```python
file = 'adult20.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
#choose column
col = 'PHSTAT_A'
col_name = 'Health Status'
freq = df[col].value_counts().rename_axis(col_name).reset_index(name = 'freq')
freq = freq.sort_values(by = col_name)
print(freq)
```

```python
#assign codebook to column
codebook = [['1', 'Excellent'],
            ['2', 'Very Good'],
            ['3', 'Good'],
            ['4', 'Fair'],
            ['5', 'Poor'],
            ['7', 'Refused'],
            ['9', 'Do Not Know']]

for lbl in codebook:
    freq[col_name] = freq[col_name].replace(int((lbl[0])), lbl[1])
    
    print(freq)
```

```python
#make frequency graph
ax = freq.plot.barh(x = col_name, y = 'freq', rot = 0)
```

```python
#choose column
col1 = 'HISPALLP_A'
col_name1 = 'Race/Hispanic Origin'
freq1 = df[col1].value_counts().rename_axis(col_name1).reset_index(name = 'freq1')
freq1 = freq1.sort_values(by = col_name1)
print(freq1)
```

```python
#assign codebook to column
codebook = [['1', 'Hispanic'],
            ['2', 'NH White Only'],
            ['3', 'NH Black/African American Only'],
            ['4', 'NH Asian Only'],
            ['5', 'NH AIAN Only'],
            ['6', 'NH AIAN and any other group'],
            ['7', 'Other single and multiple races']]

for lbl in codebook:
    freq1[col_name1] = freq1[col_name1].replace(int((lbl[0])), lbl[1])
    
    print(freq1)
```

```python
#make frequency graph
ax = freq1.plot.barh(x = col_name1, y = 'freq1', rot = 0, color = 'pink')
```
